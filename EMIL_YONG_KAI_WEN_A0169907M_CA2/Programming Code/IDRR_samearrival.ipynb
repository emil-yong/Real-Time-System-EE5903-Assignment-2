```python
def ATT_AWT_IDDR_Same_arrival_time(number_of_processes,record_of_time,record_of_process_ID,list_arrival_Burst_Time,list_arrival_time):
    print("Total Number of Tasks: {} unique tasks\n".format(number_of_processes))
    #Records of the process that has completed or left the CPU
    print(record_of_process_ID)
    # Records of time the process left the cpu 
    print(record_of_time)
    
    ATT = 0 
    array = np.array(record_of_process_ID)
    record = []
    
    cs = 0
    cs_tracker = 0
    # Getting context switching
    for i in range(len(record_of_process_ID)-1):
        cs_tracker = record_of_process_ID[i]
        cs_next = record_of_process_ID[i+1]
        if cs_next != cs_tracker:
            cs = cs+1

        # Getting Response Time 
    record_response = []
    unique_process=np.unique(record_of_process_ID)
    
    for i in (unique_process):
        #Index of the where the ID is in record_of_process ID in the list is the same as the 
        #Index in record_of_time where the ID exit the CPU
        index=np.where(record_of_process_ID == i)
        min_index = min(index[0])
        # From the list, find the minimum index because process might assign to the cpu and exit more 
        # than 1 time
        
        # Minus 1 index from the minimum because the index before the exit time's index of the located ID 
        # is the arrival time to the procecss 
        get_acutal = min_index - 1
        # for first task , index is negative for index 0 after subtraction is 0
        # set the arrival time for the first index to 0
        if get_acutal <0:
            record_response.append(0)
        else:
            response = record_of_time[get_acutal] - 0 # minus 0 cause all arrival time is the same
            record_response.append(response)
        
    
    print()
    print("The response time for each process")
    print(unique_process)
    print(record_response)
    print()
    # Finding the average response time from all processes
    average_response  = np.sum(record_response)/len(unique_process)
    
    # Finding Throughput of the CPU system
    # Throughput is the number of process completed per unit of time. 
    max_completion_time = max(record_of_time)
    min_arrival_time = 0 # because this is same arrival where all is set to 0
    throughput = number_of_processes/(max_completion_time-min_arrival_time)
    print()
    print("Throughtput for {} is {}".format(number_of_processes,throughput))

        
    
    for i in range(number_of_processes):
        task_ID = i+1
        locate = np.where(array==task_ID)
        record.append(record_of_time[max(locate[0])])
     # Calculate the average waiting time 
    AWT = np.sum((record-list_arrival_time)-list_arrival_Burst_Time)/number_of_processes
    print("\nWaiting Time of each process ID")
    print((record-list_arrival_time)-list_arrival_Burst_Time)
    print("\nTurn-Around-Time of each process ID")
    print(record-list_arrival_time)
    # Calculate the average turn around time 
    ATT = np.sum(record-list_arrival_time) /number_of_processes
    return ATT,AWT,cs,average_response,throughput
```

```python
print("########################################################")
print("Improved Round Robin Technique With Same Arrival Time")
print("########################################################\n")


# Reading the task set
dataFrame = pd.read_excel('Simulation_samearrival_30_bt_50,20.xlsx')
my_list = list(dataFrame)
print(dataFrame)
print()
print(my_list)

# Get the Process ID , arrival time and burst time 
list_process_ID = np.array(dataFrame[my_list[0]].tolist())
list_arrival_Times = np.array(dataFrame[my_list[1]].tolist())
list_arrival_Burst_Time = np.array(dataFrame[my_list[2]].tolist())

# Assertion Error means there is some missing information when manually keyed in 
assert len(list_arrival_Burst_Time) == len(list_arrival_Times) == len(list_process_ID), 'Please Check That all Process ID /Arrival Time/ Burst Time are keyed in correctly'
status_of_IDRR = None 
is_all_zero = np.all(list_arrival_Times==int(0))
if is_all_zero:
    status_of_IDRR ='Same Arrival Tasks'
else:
    status_of_IDRR = "Different Arrival Tasks"

print('\n###################################################')
print('The status of IDRR: {}'.format(status_of_IDRR))
print('###################################################\n')
# use argsort to sort the burst time and get the index

###############################################################
# The main difference between IDRR with different arrival time and 
# IDRR with same arrival is how the quantum time is set
# Note that in IDRR with different arrival time, the time quantum is 
# set differently at different interval of time.
# quantum time for first 2 iteration is different from quantum time
# for subsequents rounds
###############################################################

sorted_index = np.argsort(list_arrival_Burst_Time)

sorted_dataFrame = pd.DataFrame()
sorted_dataFrame[my_list[0]] = list_process_ID[sorted_index]
sorted_dataFrame[my_list[1]] = list_arrival_Times[sorted_index]
sorted_dataFrame[my_list[2]] = list_arrival_Burst_Time[sorted_index]

print("Sorted Based On Burst Time From Lowest To Highest\n")
print(sorted_dataFrame)

sorted_list_process_ID = list_process_ID[sorted_index]
sorted_list_arrival_Time = list_arrival_Times[sorted_index]
sorted_list_arrival_Burst_Time = list_arrival_Burst_Time[sorted_index]

Time_Quantum = (sorted_list_arrival_Burst_Time[-1] + sorted_list_arrival_Burst_Time[-2])/2

print("\nTime Quantum = {}\n".format(Time_Quantum))
sorted_dataFrame_numpy = sorted_dataFrame.to_numpy()


container_Ready_Queue = sorted_dataFrame_numpy
print(container_Ready_Queue)

T = 0
record_of_time = []
record_of_process_ID = []

remaining_processes = []

tracker = container_Ready_Queue.shape[0]
count = 0
while tracker != 0:
    number_of_process_in_ready_q= container_Ready_Queue.shape[0]
    print("\nThe Number Of Processes In Ready Queue: {}\n".format(number_of_process_in_ready_q))    
    for i in range(number_of_process_in_ready_q):
        # if burst time is lesser than TQ, assign to the CPU
        
        if container_Ready_Queue[i,2] <= Time_Quantum:
            print("Assign Process {} to CPU".format(int(container_Ready_Queue[i,0])))
            T = T + container_Ready_Queue[i,2]
            print("Value of T {}".format(T))
            record_of_process_ID.append(int(container_Ready_Queue[i,0]))
            record_of_time.append(float(T))
        else:
            # address starvation
            # still considers larger task
            record_of_process_ID.append(int(container_Ready_Queue[i,0]))
            T = T + Time_Quantum
            record_of_time.append(T)
            #update the remaining process with the remanining execution time after subtracting the QT
            remaining_processes.append([container_Ready_Queue[i,0], container_Ready_Queue[i,1], container_Ready_Queue[i,2]-Time_Quantum])
    # If there are remaining task update the TQ based on the remaining burst time
    if len(remaining_processes) > 0:
        print("\n###########################################")
        print("Remaining Process Exist")
        print("###########################################\n")
        remaining_processes = np.array(remaining_processes)
        Max_Remaining_burst = max(remaining_processes[:,2])
        print("The Max Remaining Burst Time = {}".format(Max_Remaining_burst))
        
        Time_Quantum = (Time_Quantum + Max_Remaining_burst)/2
        print("Updated  Time Quantum = {}".format(Time_Quantum))

        container_Ready_Queue = remaining_processes
        remaining_processes = []
        count = count +1
    else:
        break

print()
print("Schedule Process ID : {}".format(record_of_process_ID))
print("Schedule Burst Timer: {}".format(record_of_time))
print()
```
