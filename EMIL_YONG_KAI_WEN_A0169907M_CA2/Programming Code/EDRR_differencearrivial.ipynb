```python
def ATT_AWT_IDDR_Same_arrival_time(number_of_processes,record_of_time,record_of_process_ID,list_arrival_Burst_Time,list_arrival_time,Back_up_list_process_ID):
    print("Total Number of Tasks: {} unique tasks\n".format(number_of_processes))
    #Records of the process that has completed or left the CPU
    print(record_of_process_ID)
     # Records of time the process left the cpu 
    print(record_of_time)
    
    cs = 0
    cs_tracker = 0
    
    #Getting Context Switching
    
    print(list_arrival_time)
    for i in range(len(record_of_process_ID)-1):
        cs_tracker = record_of_process_ID[i]
        cs_next = record_of_process_ID[i+1]
        if cs_next != cs_tracker:
            cs = cs+1
    
    
    # Getting Response Time 
    record_response = []
    unique_process=np.unique(record_of_process_ID)
    #Because record_of_process_ID can have repeated process coming back to the CPU and exit
    #Find the unique ID from this list
    
    for i in (unique_process):
        index=np.where(record_of_process_ID == i)
        index_of_process = np.where(Back_up_list_process_ID==i)
        # Record_of_process_ID is recording the exit time
        # If burst time is too long, some processes might have left over
        # So there may be multiple exit time
        # Find minimum exit time of the process 
        min_index = min(index[0]) 
        # From the list, find the minimum index because process might assign to the cpu and exit more 
        # than 1 time
        
        # Minus 1 index from the minimum because the index before the exit time's index of the located ID 
        # is the arrival time to the procecss 
        get_acutal = min_index - 1 
        if get_acutal <0: # but if it is the first process, it is 0 
            record_response.append(0)
        else:
            response = record_of_time[get_acutal] - list_arrival_time[index_of_process[0][0]] 
            record_response.append(response)
    
    print()
    print("The response time for each process")
    print(unique_process)
    print(record_response)
    print()
     # Finding the average response time from all processes
    average_response  = np.sum(record_response)/len(unique_process)
    
    # Finding Throughput of the CPU system
    # Throughput is the number of process completed per unit of time. 
    max_completion_time = max(record_of_time)
    min_arrival_time = min(list_arrival_time)
    throughput = number_of_processes/(max_completion_time-min_arrival_time)
    print()
    print("Throughtput for {} is {}".format(number_of_processes,throughput))
        
    

    ATT = 0 
    array = np.array(record_of_process_ID)
    record = []
    print(number_of_processes)
    for i in range(number_of_processes):
        task_ID = i+1
        locate = np.where(array==task_ID)
        record.append(record_of_time[max(locate[0])]) # finding the exit time 
    # Calculate the average waiting time 
    AWT = np.sum((record-list_arrival_time)-list_arrival_Burst_Time)/number_of_processes
    print("\nWaiting Time of each process ID")
    print((record-list_arrival_time)-list_arrival_Burst_Time)
    print("\nTurn-Around-Time of each process ID")
    print(record-list_arrival_time)
    ATT = np.sum(record-list_arrival_time) /number_of_processes
    return ATT,AWT,cs,average_response,throughput
```

```python
print("########################################################")
print("Efficient Round Robin Technique With Different Arrival Time")
print("########################################################\n")

#########################################################

# uncomment this if you are regenerating the task set
#dataFrame = pd.read_excel('EDRR_different.xlsx')

###################################################
dataFrame = pd.read_excel('Simulation_diffarrival_30_at_10_5_bt_50_20.xlsx')
my_list = list(dataFrame)
print(dataFrame)
print()
print(my_list)

# Get the process ID , burst time , arrival time 
list_process_ID = np.array(dataFrame[my_list[0]].tolist())
list_arrival_Times = np.array(dataFrame[my_list[1]].tolist())
list_Burst_Time = np.array(dataFrame[my_list[2]].tolist())
list_Burst_Time = list_Burst_Time.astype(float)


# create a back up incase of overwrite
Back_up_list_process_ID = np.array(dataFrame[my_list[0]].tolist())
Back_up_list_arrival_Times = np.array(dataFrame[my_list[1]].tolist())
Back_up_list_Burst_Time = np.array(dataFrame[my_list[2]].tolist())



###################################################################################
# Find all the burst time that is 0
# One assumption that if all come at the same time, we should take the max 
# so if there are more than 1 task in the ready queue at time = 0, we take them max burst time
# where the Quantuam time = 0.8 * max_burst_time.

# If those that are not wihtin the QT, it will be push back to ready Q for the next iteration
# when excuting the task the CPU, other task may arrive in the ready queue.
# so algorithm will search for those arrival time that is < than the execution of the burst time
# Then it will find the 0.8* of the bursttime time of the current burst time

# it will check, the current Ready Queue , has all the process arrive?
# If yes, it will set the QT to max burst time in the next iteration
# else, it will search for all those arrival time < burst time 
# and the next burst time will still be 0.8*max_of the burst time of the current ready quee

# Actually this is addressed in setting up the dataset

#####################################################################################
Max_ID_to_track = len(list_process_ID) 
ID_tracker = 0
# this is to know that once the last process ID has arrive in the ready quee, 
# the next time it will be Qt = Max burst time 

record_of_time = []
record_of_process_ID = []

remaining_processes_id =[]
remaining_processes_arrival_time= []
remaining_processes_burst_time = []
flag = 0 
# Find all process at time = 0
while_stopper = 0



T = 0 
while while_stopper != 1:
    
    if flag == 0:
        
        index_of_zero = np.where(list_arrival_Times == 0)
        index_of_zero = index_of_zero[0]
        
        # record how many process the ready queue has encounter the process
        ID_t_id  = np.unique(list_process_ID[index_of_zero])
        #because can have left over and task coming back in to the ready q
        ID_tracker = len(ID_t_id)
        
        Burst_of_those_at_t_0  = list_Burst_Time[index_of_zero]
        Process_ID_of_those_at_t_0 = list_process_ID[index_of_zero]
        
        # find the maximum burst time to determine the quantum time 
        max_burst_of_those_at_t_0 = max(Burst_of_those_at_t_0)
        QT = 0.8*max_burst_of_those_at_t_0
        print("\nThe First Quantum Time is updated to: {}\n".format(QT))
        
        
        # get the ID, burst time and arrival time for the ready Q
        ready_queue_process_ID = list_process_ID[index_of_zero]
        ready_queue_burst_time = list_Burst_Time[index_of_zero]
        ready_queue_arrival_time = list_arrival_Times[index_of_zero]
        
        for i in range (len(ready_queue_process_ID)):
            if ready_queue_burst_time[i] <= QT:
                # If the task burst time is lesser than the QT, send to the CPU
                # and record down the execution time (exit time or completion time) and process ID
                T = T + ready_queue_burst_time[i]
                record_of_process_ID.append(ready_queue_process_ID[i])
                record_of_time.append(T)
                
                #once that task has been send to the CPU can remove form the ready q
                index_of_interest = np.where(list_process_ID == ready_queue_process_ID[i])
                list_process_ID=np.delete(list_process_ID,index_of_interest,0)
                list_arrival_Times=np.delete(list_arrival_Times,index_of_interest,0)
                list_Burst_Time=np.delete(list_Burst_Time,index_of_interest,0)
                
            else:
                # update the QT for larger task
                remainder_burst = ready_queue_burst_time[i] - QT
                T = T + QT
                
                index_of_remainer = np.where(list_process_ID == ready_queue_process_ID[i])
                list_Burst_Time[index_of_remainer] = remainder_burst
                
                record_of_process_ID.append(ready_queue_process_ID[i])
                record_of_time.append(T)
                
                remaining_processes_id.append(ready_queue_process_ID[i])
                remaining_processes_burst_time.append(remainder_burst)
                remaining_processes_arrival_time.append(ready_queue_arrival_time[i])
        # the container that consist of large task set
        conatiner_remaining_processes = list(zip(list_process_ID, list_arrival_Times,list_Burst_Time))
        df = pd.DataFrame (conatiner_remaining_processes, columns = ['Process ID','Arrival Time','Burst Time'])
        print(df)
        print()
        print("After 1st iteration dataset ")
        
        # set the ready queue to handle those remaining process so the loops
        # repeats again
        ready_queue_process_ID = remaining_processes_id
        ready_queue_arrival_time=remaining_processes_arrival_time
        ready_queue_burst_time=remaining_processes_burst_time
        
        remaining_processes_id=[]
        remaining_processes_arrival_time=[]
        remaining_processes_burst_time=[]
        
        flag=1 # set the flag so it can go to the else 
    else:
        ##
        # Consider if there are more task arriving
        # this is account for those task arriving while task executing 
        ##
        index_of_those_arriving= [i for i,v in enumerate(list_arrival_Times) if v <= T]        
        print(index_of_those_arriving)
        
        ID_t_id = np.append(ID_t_id, list_process_ID[index_of_those_arriving])
        ID_t_id  = np.unique(ID_t_id)
        
        print(ID_t_id)

        # prepare the ready queue
        # need consider the left over from the t = 0
        for i in range(len(index_of_those_arriving)):
            if list_process_ID[index_of_those_arriving[i]] in ready_queue_process_ID:
                pass
            else:
                ready_queue_process_ID.append(list_process_ID[index_of_those_arriving[i]])
                ready_queue_arrival_time.append(list_arrival_Times[index_of_those_arriving[i]])
                ready_queue_burst_time.append(list_Burst_Time[index_of_those_arriving[i]])
        # same as first iteration EDRR
        max_burst_time_in_list = max(ready_queue_burst_time)
        QT = 0.8*max_burst_time_in_list
        print("\nThe new Quantum Time is updated to: {}\n".format(QT))
        for i in range(len(ready_queue_burst_time)):
            if ready_queue_burst_time[i] <= QT:
                T = T + ready_queue_burst_time[i]
                record_of_process_ID.append(ready_queue_process_ID[i])
                record_of_time.append(T)
                index_of_interest = np.where(list_process_ID == ready_queue_process_ID[i])
                list_process_ID=np.delete(list_process_ID,index_of_interest,0)
                list_arrival_Times=np.delete(list_arrival_Times,index_of_interest,0)
                list_Burst_Time=np.delete(list_Burst_Time,index_of_interest,0)
            else:
                remaining_processes_id.append(ready_queue_process_ID[i])
                remaining_processes_burst_time.append(ready_queue_burst_time[i])
                remaining_processes_arrival_time.append(ready_queue_arrival_time[i])

        conatiner_remaining_processes = list(zip(list_process_ID, list_arrival_Times,list_Burst_Time))
        df = pd.DataFrame (conatiner_remaining_processes, columns = ['Process ID','Arrival Time','Burst Time'])
        print()
        print(df)

        ready_queue_process_ID = remaining_processes_id
        ready_queue_arrival_time=remaining_processes_arrival_time
        ready_queue_burst_time=remaining_processes_burst_time

        remaining_processes_id=[]
        remaining_processes_arrival_time=[]
        remaining_processes_burst_time=[]
        # if all the task has arrive, QT is set to the max of the remaining task
        if len(ID_t_id) == Max_ID_to_track:
            print("All processes have arrived in the ready queue")

            QT = max(ready_queue_burst_time)
            print("\nThe new Quantum Time is updated to: {}\n".format(QT))          

            for i in range(len(ready_queue_process_ID)):
                if ready_queue_burst_time[i] <= QT:
                    T = T + ready_queue_burst_time[i]
                    record_of_process_ID.append(ready_queue_process_ID[i])
                    record_of_time.append(T)
                else:
                    print("error if this happens")

            print("All have already been assigned to the CPU ........")
            print(record_of_process_ID)
            print(record_of_time)
            while_stopper = 1
        else:
            # there are still processes available that have not arrive so we need to continue and let it search more
            print("There are still other process not yet arrive")
            pass 



```
