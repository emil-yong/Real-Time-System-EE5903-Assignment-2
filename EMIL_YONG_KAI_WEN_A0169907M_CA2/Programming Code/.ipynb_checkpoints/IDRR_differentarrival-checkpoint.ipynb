```python
def ATT_AWT_IDDR_Same_arrival_time(number_of_processes,record_of_time,record_of_process_ID,list_arrival_Burst_Time,list_arrival_time,list_process_ID):
    print("Total Number of Tasks: {} unique tasks\n".format(number_of_processes))
        #Records of the process that has completed or left the CPU
    print(record_of_process_ID)
        # Records of time the process left the cpu 
    print(record_of_time)
    
    ATT = 0 
    array = np.array(record_of_process_ID)
    record = []
    
    cs = 0
    cs_tracker = 0
       # Getting context switching
    for i in range(len(record_of_process_ID)-1):
        cs_tracker = record_of_process_ID[i]
        cs_next = record_of_process_ID[i+1]
        if cs_next != cs_tracker:
            cs = cs+1
    

        # Getting Response Time 
    record_response = []
    unique_process=np.unique(record_of_process_ID)
    
    for i in (unique_process):
        #Index of the where the ID is in record_of_process ID in the list is the same as the 
        #Index in record_of_time where the ID exit the CPU
        index=np.where(record_of_process_ID == i)
        index_of_process = np.where(list_process_ID==i)
        min_index = min(index[0])
        # From the list, find the minimum index because process might assign to the cpu and exit more 
        # than 1 time
        
        # Minus 1 index from the minimum because the index before the exit time's index of the located ID 
        # is the arrival time to the procecss 
        get_acutal = min_index - 1
        # for first task , index is negative for index 0 after subtraction is 0
        # set the arrival time for the first index to 0
        if get_acutal <0:
            record_response.append(0)
        else:
            response = record_of_time[get_acutal] - list_arrival_time[index_of_process[0][0]] # minus 0 cause all arrival time is the same
            record_response.append(response)
    
    print()
    print("The response time for each process")
    print(unique_process)
    print(record_response)
    print()
    
    average_response  = np.sum(record_response)/len(unique_process)
    
    # Finding Throughput of the CPU system
    # Throughput is the number of process completed per unit of time. 
    max_completion_time = max(record_of_time)
    min_arrival_time = min(list_arrival_time)
    throughput = number_of_processes/(max_completion_time-min_arrival_time)
    print()
    print("Throughtput for {} is {}".format(number_of_processes,throughput))
        
    
    for i in range(number_of_processes):
        task_ID = i+1
        locate = np.where(array==task_ID)
        record.append(record_of_time[max(locate[0])])
    # Calculate the average waiting time 
    AWT = np.sum((record-list_arrival_time)-list_arrival_Burst_Time)/number_of_processes
    print("\nWaiting Time of each process ID")
    print((record-list_arrival_time)-list_arrival_Burst_Time)
    print("\nTurn-Around-Time of each process ID")
    print(record-list_arrival_time)
    # Calculate the average turn around time 
    ATT = np.sum(record-list_arrival_time) /number_of_processes
    return ATT,AWT,cs,average_response,throughput
```

```python
# print("########################################################")
print("Improved Round Robin Technique With Different Arrival Time")
print("########################################################\n")


# Reading the task set
dataFrame = pd.read_excel('Simulation_diffarrival_30_at_10_5_bt_50_20.xlsx')
my_list = list(dataFrame)
print(dataFrame)
print()
print(my_list)

# Get the Process ID , arrival time and burst time 
list_process_ID = np.array(dataFrame[my_list[0]].tolist())
list_arrival_Times = np.array(dataFrame[my_list[1]].tolist())
list_Burst_Time = np.array(dataFrame[my_list[2]].tolist())

# Assertion Error means there is some missing information when manually keyed in 
assert len(list_Burst_Time) == len(list_arrival_Times) == len(list_process_ID), 'Please Check That all Process ID /Arrival Time/ Burst Time are keyed in correctly'
status_of_IDRR = None 
is_all_zero = np.all(list_arrival_Times==int(0))

if is_all_zero:
    status_of_IDRR ='Same Arrival Tasks'
else:
    status_of_IDRR = "Different Arrival Tasks"
    
print('\n###################################################')
print('The status of IDRR: {}'.format(status_of_IDRR))
print('###################################################\n')

###############################################################
# The main difference between IDRR with different arrival time and 
# IDRR with same arrival is how the quantum time is set
# Note that in IDRR with different arrival time, the time quantum is 
# set differently at different interval of time.
# quantum time for first 2 iteration is different from quantum time
# for subsequents rounds
###############################################################


T = 0
record_of_time = []
record_of_process_ID = []
conatiner_remaining_processes = []

# get a container for the ID , arrival time and burst time 
container_list_process_ID = np.array(dataFrame[my_list[0]].tolist())
container_list_arrival_Times = np.array(dataFrame[my_list[1]].tolist(),dtype=float)
container_list_Burst_Time = np.array(dataFrame[my_list[2]].tolist(),dtype=float)


# =======================
# Find the arrival time which is the smallest 
first_smallest_arrival_time = min(list_arrival_Times)
find_all_index = np.where(list_arrival_Times==first_smallest_arrival_time)
# randomly pick one in the list 
# for simplicity will pick the first one if there are more than 1 tasks in the ready queue
# right at the begining and get the burst time

First_task_index= find_all_index[0][0]

First_task_process_id = list_process_ID[First_task_index]
print("The First Process ID to be assigned to the CPU is {}\n".format(First_task_process_id))
First_task_arrival_time = list_arrival_Times[First_task_index]
QT = list_Burst_Time[First_task_index]

# set the QT to the first burst time
# see appendix for the pseudocode or case 3 in the academic paper
First_QT= QT 
print("The First Quantum time is {}\n".format(QT))

# Record the T for ghantt chart
T = T + QT
# Assign first task to the CPU and record the exit time and the ID for ghantt chart
record_of_time.append(T)
record_of_process_ID.append(First_task_process_id)

# remove the task that has been assigned to the CPU
container_list_process_ID =(np.delete(container_list_process_ID,First_task_index,0))
container_list_arrival_Times = (np.delete(container_list_arrival_Times,First_task_index,0))
container_list_Burst_Time = (np.delete(container_list_Burst_Time,First_task_index,0))

conatiner_remaining_processes = list(zip(container_list_process_ID, container_list_arrival_Times, container_list_Burst_Time))

df = pd.DataFrame (conatiner_remaining_processes, columns = ['Process ID','Arrival Time','Burst Time'])

# Pay attention that at time = 0, there are also other tasks
# Might have more than 1 task at time = 0
# we only just chooose the first task and set the time quantum based on the pseudocode

length_of_find_all_index = len(find_all_index[0])

for i in range(length_of_find_all_index-1):
    i = i +1 # we do not want to consider the first index since we already used it for the QT
    index_of_interest = find_all_index[0][i]
    
    process_id_of_interest = list_process_ID[index_of_interest]
    burst_time_of_interest = list_Burst_Time[index_of_interest]
    
    # assign the rest of the task that has the burst time < QT
    # record the exit time and id for ghantt chart
    if burst_time_of_interest <=QT:
        record_of_process_ID.append(process_id_of_interest)
        T = T + burst_time_of_interest
        record_of_time.append(T)
    else:
        # still considers the large task
        # address starvation
        # record the exit time and ID
        # since got remaining execution time, add back to container
        remaining_burst_time = burst_time_of_interest - QT
        record_of_process_ID.append(process_id_of_interest)
        T = T + QT
        record_of_time.append(T)
        
        index_of_ID = np.where(container_list_process_ID == process_id_of_interest)
        container_list_Burst_Time[index_of_ID] = remaining_burst_time
    
conatiner_remaining_processes = list(zip(container_list_process_ID, container_list_arrival_Times, container_list_Burst_Time))
df = pd.DataFrame (conatiner_remaining_processes, columns = ['Process ID','Arrival Time','Burst Time'])
print("Update the ready queue after 1st iteration\n")
print(df)
print("##########################################")
tracker = 0
count_temp = 0
whilebreaker=0
while len(container_list_arrival_Times)!= 0:  
    ## during the excution of the first
    container_ready_que = []
    
    # this is account for those task arriving while executing the first time 
    index_of_those_arriving= [i for i,v in enumerate(container_list_arrival_Times) if v <= T]
    
    #get the id , arrrival , burs time for ready Q
    ready_que_id = container_list_process_ID[index_of_those_arriving]
    ready_que_arrival = container_list_arrival_Times[index_of_those_arriving]
    ready_que_bursttime = container_list_Burst_Time[index_of_those_arriving]

    
    # sort the ready queue
    sort_ready_que_index = np.argsort(ready_que_bursttime)
    ready_que_id = ready_que_id[sort_ready_que_index]
    ready_que_arrival=ready_que_arrival[sort_ready_que_index]
    ready_que_bursttime = ready_que_bursttime[sort_ready_que_index]
    
    if tracker == 0:
        # update the TQ for round 1
        # note this TQ is updated ONCE!
        high_burst = max(ready_que_bursttime)
        old_qt = QT
        unique_arrival = np.unique(ready_que_arrival)
        ar_one = unique_arrival[0]
        ar_two = unique_arrival[1]
        low_ar = (ar_one+ar_two)/2
        QT = ((high_burst+old_qt)/2) - low_ar 
        tracker = 1
        print("\nUpdate of the Quantum Time once {}\n".format(QT))
        
        

    else:
        print("##########################################")
        # after round 1, subseqeunt TQ is updated this way
        high_burst = max(ready_que_bursttime)
        old_qt = QT
        index_ar = np.argsort(ready_que_arrival)
        low_ar = ready_que_arrival[index_ar[0]]/2
        
        QT = ((high_burst+old_qt)/2) - low_ar
        
        print("\nTime Quantum after third iteration onward {}\n".format(QT))

    ## check if the task is within the new Quantum time 
    for i in range(len(ready_que_id)):
        # assign to CPU
        # and record the ID and exit time 
        if ready_que_bursttime[i] <= QT:
            record_of_process_ID.append(ready_que_id[i])
            T = T + ready_que_bursttime[i]
            record_of_time.append(T)
            index_of_interest = np.where(container_list_process_ID == ready_que_id[i])
            container_list_arrival_Times=np.delete(container_list_arrival_Times,index_of_interest,0)
            container_list_process_ID=np.delete(container_list_process_ID,index_of_interest,0)
            container_list_Burst_Time=np.delete(container_list_Burst_Time,index_of_interest,0)
            
        else:

            remaining_burst_time = float(ready_que_bursttime[i] - QT)

            record_of_process_ID.append(ready_que_id[i])
            T = T + QT
            record_of_time.append(T)
            
            index_of_ID = np.where(container_list_process_ID == ready_que_id[i])
 
            container_list_Burst_Time[index_of_ID] = float(remaining_burst_time)
    
    if len(container_list_arrival_Times) !=0 :
        conatiner_remaining_processes = list(zip(container_list_process_ID, container_list_arrival_Times, container_list_Burst_Time))
        df = pd.DataFrame (conatiner_remaining_processes, columns = ['Process ID','Arrival Time','Burst Time'])
        print("Update the ready queue with the Time Quantum ")
        print()
        print(df)

print("All assigned to CPU already............\n")
print("Record Of Process to CPU")
print(record_of_process_ID)
print('\nRecord of time')
print(record_of_time)

```
